```python
#https://www.youtube.com/watch?v=q8V6rv8gr8o
from simpletransformers.language_representation import RepresentationModel
        
sentences = ["Machine Learning and Deep Learning are part of AI", "Data Science will excel in future"] #it should always be a list
model = RepresentationModel(
        model_type="bert",
        model_name="bert-base-uncased",
        use_cuda=True
    )
#word embedding
word_vectors = model.encode_sentences(sentences, combine_strategy=None)
print(word_vectors.shape) 
print('first sentence is {}'.format(word_vectors))

# #sentence embedding
# sentence_vectors = model.encode_sentences(sentences, combine_strategy="mean")
# print(sentence_vectors.shape)
# print(word_vectors)
```

Another Approach

```python
# implementing by using tensorflow
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained('bert-base-uncased')
input_ids = tf.constant(tokenizer.encode("Hello, my dog is cute"))[None, :]  # Batch size 1
outputs = model(input_ids)
last_hidden_states = outputs[0]  # The last hidden-state is the first element of the output tuple

print(tokenizer.tokenize("Hello, my dog is cute"))
print(last_hidden_states[0].shape) # the first and the last ones are the cls and sep
```

```python
## This proves that if we have same sentence in two different documents, the embedding for it will be different. That means this approach works.  
sentences1 = "Russia said it was continuing to pull troops back from near Ukraine, but Secretary of State Antony Blinken and the head of NATO said they had seen no indication of a withdrawal. Still, Western officials were cautiously hopeful for a diplomatic solution to the crisis. Machine Learning and Deep Learning are part of AI. Hello, my dog is cute"
input_ids1 = tf.constant(tokenizer.encode(sentences1))[None, :]  # Batch size 1
outputs1 = model(input_ids1)
last_hidden_states1 = outputs1[0]
dict1 = dict()
for el, val in zip(last_hidden_states[0], tokenizer.tokenize("Hello, my dog is cute")):
  dict1[val] = el

dict2 = dict()
for el, val in zip(last_hidden_states1[0], tokenizer.tokenize(sentences1)):
  dict2[val] = el

import numpy as np
vec1 = np.array(dict1['dog'])
vec2 = np.array(dict2['dog'])
total = 0
for x, y in zip(vec1, vec2):
  total += (x-y)**2
print(total)

```
